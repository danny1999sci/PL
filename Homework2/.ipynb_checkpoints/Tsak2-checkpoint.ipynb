```python
import numpy as np 
import matplotlib.pyplot as plt 

import pandas as pd
df = pd.read_csv('Starbucks satisfactory survey.csv')
```

#統計男生以及女生 每天的消費金額
#性別為男性或女性

```python
male_data = df.loc[df['Gender'] == 'Male']
female_data = df.loc[df['Gender'] == 'Female']
```

#每個性別的數據按日期進行分組，並對每個組計算消費金額的總和

```python
male_daily_sales = male_data.groupby('Timestamp')['AannualIncome'].sum()
female_daily_sales = female_data.groupby('Timestamp')['AannualIncome'].sum()
```

```python
import matplotlib.pyplot as plt

plt.plot(male_daily_sales.index, male_daily_sales.values, label='Male')
plt.plot(female_daily_sales.index, female_daily_sales.values, label='Female')
plt.xlabel('Timestamp')
plt.ylabel('Total sales')
plt.legend()
plt.show()
```
