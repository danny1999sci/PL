```python
import pandas as pd
import matplotlib.pyplot as plt

# 加载数据集
data = pd.read_csv('2015_Air_quality_in_northern_Taiwan.csv')

# 将日期时间列转换为datetime对象
data['date'] = pd.to_datetime(data['time'])

# 从日期时间列中提取年份、月份和日期
data['year'] = data['date'].dt.year
data['month'] = data['date'].dt.month
data['day'] = data['date'].dt.day

# 计算每个月份的PM2.5平均值
pm25_monthly = data.groupby('month')['PM2.5'].mean()

# 绘制折线图显示PM2.5月均值的季节性变化
plt.plot(pm25_monthly)
plt.title('PM2.5月均值季节性变化')
plt.xlabel('月份')
plt.ylabel('PM2.5月均值')
plt.show()
#这段代码会绘制一个折线图，显示每个月份的PM2.5平均值，以展示空气质量是否有季节性变化。如果图形显示的PM2.5值在不同月份之间存在明显差异，则可以得出结论，空气质量在不同季节之间有所变化。





```
